```python
# List of customer orders (fruits)
customer_value=["mango","dates","apple","pineapple","apple","dates","mango","dates","apple"]
# This list will store items we have already counted
checked=[]
# Outer loop: go through each item in customer_value
for i in range(9):
    item=customer_value[i] # current fruit to count
      # Check if we have already counted this item
    if item  not in checked:
        count=0 # start counting from 0
        for j in range(9):# Inner loop: check every item in the list to count occurrences
            if item==customer_value[j]:
                count=count+1
                 # After counting all occurrences, print the item and its count
        print(item,"=",count)
        checked.append(item)
       
```

```python
customer_value=["mango","dates","apple","pineapple","apple","dates","mango","dates","apple"]
checked=[]
max_count=0
most_item=""
for i in range(9):
    item=customer_value[i]
    if item not in checked:
        count=0
        for j in range(9):
            if item==customer_value[j]:
                count=count+1
        print(item,"=",count)
        checked.append(item)
        if count>max_count:
            max_count=count
            most_item=item
        checked.append(item)

print("Most ordered:", most_item) 
            
```
